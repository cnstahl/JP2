```python
N=10
indices = np.zeros((2**N,2**N))
for i in range(2**N):
    for j in range(2**N):
        if Fermi(j,10) == Fermi(i,10): indices[i,j] = 10
plt.imshow(np.log(indices), cmap='YlOrRd', interpolation='nearest')
```
